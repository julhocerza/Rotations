```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
m = 1
r = 1
omega = 1
```

```python
t_array = np.linspace(0, 2*np.pi, 100)
x_array = r*np.cos(omega * t_array)
y_array = r*np.sin(omega * t_array)
```

```python
fig = plt.figure()
ax = fig.add_subplot(1,1,1)
ax.set_aspect(1)

plt.xlabel('x')
plt.ylabel('y')
plt.plot(x_array, y_array)
```

```python
plt3D = plt.axes(projection='3d')
plt3D.set_xlabel('x [m]')
plt3D.set_ylabel('y [m]')
plt3D.set_zlabel('t [s]')
plt3D.scatter3D(x_array, y_array, t_array)
```

```python
n1points = 30
coord_list = []
counter = 0
contrib = 0
for x in np.linspace(-r,r,n1points):
    for y in np.linspace(-r,r,n1points):
        for z in np.linspace(-r,r,n1points):
            if(np.linalg.norm([x,y,z]) <= r):
                coord_list.append([x,y,z])
                counter = counter + 1
                contrib = contrib + np.linalg.norm([x,y,0])**2
```

```python
counter
```

```python
coord_list = np.transpose(coord_list)
```

```python
plt3D = plt.axes(projection='3d')
plt3D.set_xlabel('x [m]')
plt3D.set_ylabel('y [m]')
plt3D.set_zlabel('z [m]')
plt3D.scatter3D(coord_list[0], coord_list[1], coord_list[2],s=0.1)
```
