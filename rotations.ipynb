```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
m = 1
r = 1
omega = 1
```

```python
t_array = np.linspace(0, 2*np.pi, 100)
x_array = r*np.cos(omega * t_array)
y_array = r*np.sin(omega * t_array)
```

```python
fig = plt.figure()
ax = fig.add_subplot(1,1,1)
ax.set_aspect(1)

plt.xlabel('x')
plt.ylabel('y')
plt.plot(x_array, y_array)
```

```python
plt3D = plt.axes(projection='3d')
plt3D.set_xlabel('x [m]')
plt3D.set_ylabel('y [m]')
plt3D.set_zlabel('t [s]')
plt3D.scatter3D(x_array, y_array, t_array)
```

```python
s = 1
```

```python
npoints = 100
r_list = np.linspace(0,s,npoints)

fig = plt.figure()
ax = fig.add_subplot(1,1,1)
ax.set_aspect(1)

plt.xlabel('x')
plt.ylabel('y')
plt.xlim([-1.2*s,1.2*s])
plt.ylim([-1.2*s,1.2*s])
plt.plot(r_list, np.zeros(npoints))
```

```python
1/3 * m*s**2 / 2 * omega**2 #analitic energy
```

```python
(m/npoints)/2 * omega**2 * np.sum(r_list**2) #numerical energy
```
